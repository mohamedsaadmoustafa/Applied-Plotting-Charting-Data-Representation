```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0, 2*np.pi, 1000)
r = 5 - 5 * np.sin(theta)
```

```python
plt.figure(figsize=(14, 10))
plt.polar(theta, r, color='#791B1B');
```

```python
plt.figure(figsize=(14, 10))
plt.polar(theta, r, color='pink', lw='2');
plt.fill_between(theta, r, 0, alpha=0.5, color='r');

ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
```
