```python
import math
import sympy
import numpy as np
import matplotlib.pyplot as plt


plt.style.use('dark_background')
```

```python
class Primes:
    def __init__(self):
        self.primes = []

    def func(self, number):
        if number == 0 or number == 1: return 0

        self.primes = [1]*(number) # list of true values with length equal to input number
        self.primes[0] = self.primes[1] = 0

        p = 2 
        while (p**2 <= number):
            # If prime[p] is not changed, then it is a prime
            if self.primes[p] == 1:
                # Update all multiples of p
                for idx in range(p**2, number, p):
                    #print(p, '--', idx)
                    self.primes[idx] = 0
            p += 1
        __primes = [index for index, prime in enumerate(self.primes) if prime]
        self.primes = [] # reset index
        return __primes

Primes().func(110) == list(sympy.primerange(0, 110))
```

Create function to plot list of numbers with coordinates 

    x-coordinate = nums * np.cos(nums)
    y-coordinate = nums * np.sin(nums)

```python
def create_plot(numbers, figsize=12, s=5, show_annot=False, c='g'):
    numbers = np.array(list(numbers))
    x, y = numbers*np.cos(numbers), numbers*np.sin(numbers)
    plt.figure(figsize=(figsize, figsize))
    plt.axis("off")
    plt.scatter(x, y, s=s, c=c)
    if show_annot:
        for index, number in enumerate(numbers):
            plt.annotate(number, (x[index], y[index]))
    plt.show()

nums = range(100)
create_plot(nums, show_annot=True)
```

```python
primes = Primes().func(1000)
create_plot(primes)
```

Plot prime numbers from 0 to 100

```python
nums = Primes().func(100)
create_plot(nums, show_annot=True)
```

```python
primes = Primes().func(20000)
create_plot(primes, s=1)
```
