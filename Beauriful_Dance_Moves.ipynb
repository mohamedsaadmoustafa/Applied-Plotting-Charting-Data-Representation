```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
```

```python
y_constant = np.array([0]*len(x))

plt.scatter(0, 0, c=20)
plt.plot(x, y_constant, label='constant');
```

```python
y_linear = x.copy()

plt.scatter(0, 0, c=20)
plt.plot(x, y_linear, label='linear');
```

```python
y_absolute = np.abs(x)

plt.scatter(0, 0, c=20)
plt.plot(x, y_absolute, label='absolute');
```

```python
y_quadratic = np.power(x, 2)

plt.scatter(0, 0, c=20)
plt.plot(x, y_quadratic, label='quadratic');
```

```python
y_sqrt = np.sqrt(x)

plt.scatter(0, 0, c=20)
plt.plot(x, y_sqrt, label='sqrt');
```

```python
y_cubic = np.power(x, 3)

plt.scatter(0, 0, c=20)
plt.plot(x, y_cubic, label='cubic');
```

```python
y_cubic_root =  np.cbrt(x)

plt.scatter(0, 0, c=20);
plt.plot(x, y_cubic_root, label='cubic root');
```

```python
y_inverse = 1/x;

plt.scatter(0, 0, c=20);
plt.plot(x, y_inverse, label='inverse');
```

```python
y_rational = 1/x**2

plt.scatter(0, 0, c=20);
plt.plot(x, y_rational, label='rational');
```

```python
y_log = np.log(x)

plt.scatter(0, 0, c=20);
plt.plot(x, y_log, label='logarithmic');
```

```python
y_e = np.exp(x)

plt.scatter(0, 0, c=20);
plt.plot(x, y_e, label='exponential');
```

```python
y_sin =  np.sin(x)

plt.scatter(0, 0, c=20);
plt.plot(x, y_sin, label='Sin');
```

```python
y_cos = np.cos(x)

plt.scatter(0, 0, c=20);
plt.plot(x, y_cos, label='Sin');
```

```python
y_tan = np.tan(x)

plt.scatter(0, 0, c=20);
plt.plot(x, y_tan, label='Tan');
```

```python
y_cot = 1/y_tan

plt.scatter(0, 0, c=20);
plt.plot(x, y_cot, label='Cot');
```
