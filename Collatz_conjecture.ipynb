```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class CollatzConjecture:
    def __init__(self):
        self.dp = []
        
    def is_odd(self, number):
        if number % 2 == 0: return 0
        else:               return 1

    def func(self, number):
        if number < 1: return []
        odd = self.is_odd(number)

        if number in self.dp[:-1]: 
            index = self.dp.index(number) 
            print(f'{len(self.dp)-1} steps! \nThe Loop will continue in {self.dp[index:]} for ever!')  ##self.func(self.dp[index + 1])
            return self.dp[:-1]

        if odd: self.dp += [number * 3 + 1]
        else:   self.dp += [int(number / 2)]

        #print(self.dp)
        return self.func(self.dp[-1])
```

```python
l =  CollatzConjecture().func(10)
plt.plot(l)
```

```python
l =  CollatzConjecture().func(100)
plt.plot(l)
```

```python
l = CollatzConjecture().func(2000)
plt.plot(l)
```

```python
plt.figure(figsize=(16, 12))
for n in range(3, 20):
    c = CollatzConjecture().func(n)
    plt.plot(
        sum(
            zip( c, c[1:]),
            tuple()
        ), 
        lw = 2, 
        alpha = 0.8, 
        label = "n = %d"%n
    )
plt.grid()
plt.legend();
```
